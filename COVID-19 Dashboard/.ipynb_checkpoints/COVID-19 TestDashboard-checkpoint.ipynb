```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
from IPython.core.display import display, HTML

import pandas as pd
import numpy as np
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import folium
import plotly.graph_objects as go
import seaborn as sns
import ipywidgets as widgets
```

```python
confirmed_info = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
death_info = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recovered_info = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
country_info = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/web-data/data/cases_country.csv')
```

```python
#data cleaning process

confirmed_info.columns = map(str.lower, confirmed_info.columns)
recovered_info.columns = map(str.lower, recovered_info.columns)
death_info.columns = map(str.lower, death_info.columns)
country_info.columns = map(str.lower, country_info.columns)
```

```python
confirmed_info = confirmed_info.rename(columns={'province/state': 'state', 'country/region' : 'country'})
recovered_info = recovered_info.rename(columns={'province/state': 'state', 'country/region' : 'country'})
death_info = death_info.rename(columns={'province/state': 'state', 'country/region' : 'country'})
country_info = country_info.rename(columns={'country_region' : 'country'})
```

```python
country_info.head(5)
```

```python
# total number of confirmed, death and recovered cases
confirmed_total = int(country_info['confirmed'].sum())
deaths_total = int(country_info['deaths'].sum())
recovered_total = int(country_info['recovered'].sum())
active_total = int(country_info['active'].sum())
```

```python
# displaying the total stats

display(HTML("<div style = 'background-color: #504e4e; padding: 30px '>" +
             "<span style='color: #fff; font-size:30px;'> Confirmed: "  + str(confirmed_total) +"</span>" +
             "<span style='color: red; font-size:30px;margin-left:20px;'> Deaths: " + str(deaths_total) + "</span>"+
             "<span style='color: lightgreen; font-size:30px; margin-left:20px;'> Recovered: " + str(recovered_total) + "</span>"+
             "</div>")
       )
```

```python
#Styling the Dataframe
def styling_column(x):
    o='background-color: orange'
    r='background-color: red'
    g='background-color: green'
    temp_df = pd.DataFrame('', index = x.index, columns = x.columns)
    temp_df.iloc[:,4] = o
    temp_df.iloc[:,5] = r
    temp_df.iloc[:,6] = g
    return temp_df

country_info.style.apply(styling_column, axis=None)
```

```python
# sorting the values by confirmed descednding order
country_info.sort_values('confirmed', ascending= False).head(10).style.background_gradient(cmap='copper')
fig = go.FigureWidget( layout=go.Layout() )
def highlight_col(x):
    o = 'background-color: orange'
    y = 'background-color: yellow'
    r = 'background-color: red'
    df1 = pd.DataFrame('', index=x.index, columns=x.columns)
    df1.iloc[:, 4] = y
    df1.iloc[:, 5] = r
    df1.iloc[:, 6] = g
    
    return df1

def show_latest_cases(n):
    n = int(n)
    return country_info.sort_values('confirmed', ascending= False).head(n).style.apply(highlight_col, axis=None)

interact(show_latest_cases, n='10')

ipywLayout = widgets.Layout(border='solid 2px green')
ipywLayout.display='none' 
widgets.VBox([fig], layout=ipywLayout)
```
